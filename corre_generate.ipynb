```python
topic_member_sum = []
topic_corr = []

for i in range(len(topic_lst) - 1):
    for j in range(len(topic_lst) - i - 1):
#         print("___start___")
#         print(topic_lst[i])
#         print(topic_lst[j + i + 1])
#         print(data_corre[i][j + i + 1])
        if (topic_member_lst[i] > 50 and topic_member_lst[j + i + 1] > 50) and (data_corre[i][j + i + 1] < 0.5) :
        
            topic_member_sum.append(topic_member_lst[i] + topic_lst[j + i + 1])
            topic_corr.append(data_corre[i][j + i + 1] ** 0.25)

        
```

```python
import seaborn as sns
import numpy as np
from numpy.random import randn
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats
 
 
# style set 这里只是一些简单的style设置
sns.set_palette('deep', desat=.6)
sns.set_context(rc={'figure.figsize': (8, 5) } )
np.random.seed(1425)
# figsize是常用的参数.
sns.set()

# x = stats.gamma(2).rvs(5000)
# y = stats.gamma(50).rvs(5000)
x = np.array(topic_corr)
y = np.array(topic_member_sum)
with sns.axes_style("dark"):
    sns.jointplot(x, y, kind="hex")
    
plt.ylim(0, 400000)
plt.savefig("joindis.png", dpi=100)
```

```python
from scipy.stats import norm
g = sns.kdeplot(x, shade=True)
# g = sns.distplot(x, shade=True)
plt.xlim(0,0.2)
# g.set(xlim=(0, 0.4))
# g.fig.set_dpi(300)
ax = plt.subplot(111)
ax.set_title("Distribution of Correltion Index")
plt.savefig("corrkde.png", dpi=300)
```

```python
g = sns.kdeplot(y, shade=True)
g.set(xlim=(0, 500000))
ax = plt.subplot(111)
ax.set_title("Distribution of Topic Member Number")
plt.savefig("memberkde.png", dpi=300)
```
