```python
topic_corr = np.zeros(1623*1623).reshape(1623,1623)
topic_corr.shape
```

```python
df_dup = df_mem_topic.drop_duplicates()
topic_lst = df_dup['topic_id'].tolist()
```

```python
group_topic = df_mem_topic.groupby(by=['topic_id'])
topic_member_lst = group_topic.size().to_frame('size')['size'].tolist()
```

## This Step 

```python
for i in range(1623):
#     print(i)
    topic_a = topic_lst[i]
    for j in range(1623 - i):
        print("_____START_____")
        print(i)
        print (j + i)
        topic_b = topic_lst[j + i]
        df_mem_topic[df_mem_topic.topic_id == topic_a]
        common_member_cnt = len(pd.merge(df_mem_topic[df_mem_topic.topic_id == topic_a], df_mem_topic[df_mem_topic.topic_id == topic_b], how='inner', on=['member_id']))
#         print(common_member_cnt)
        cor_val = common_member_cnt * common_member_cnt /(topic_member_lst[i] * topic_member_lst[j + i])
#         print(cor_val)
        print("_____END_____")
        topic_corr[i][j + i] = cor_val
np.save("corr_result.npy", topic_corr)
```
