```python
def getEvents(groups):
    events = []
    for i in range(10):
        try:
            result = groups.results[i]
            temp = client.GetEvents(group_id=result['id'])
            for m in temp.results:
                events.append(m)
        except Exception:
            pass
    
    return events
```

```python
art_events = getEvents(art_groups)
business_events = getEvents(business_groups)
lgbt_events = getEvents(lgbt_groups)
language_events = getEvents(language_groups)
sport_events = getEvents(sport_groups)
tech_events = getEvents(tech_groups)
```

```python
total = 0
count = 0
for b in language_events:
    if 'duration' in b:
        total = total + b['duration']
        count = count + 1

print(total / count / 60000)
```

```python
# t_series[t_series['duration']/60000 <= 30]
event_groups = [language_events, tech_events, sport_events, lgbt_events, art_events, business_events]
total = 0
count = 0
max = 0
min = 10000000000
for eg in event_groups:
    for e in eg:
        if 'duration' in e:
            total = total + e['duration']
            count = count + 1
            if e['duration'] > max:
                max = e['duration']
            if e['duration'] < min:
                min = e['duration']

print(total / count / 60000)
print(max / 60000)
print(min / 60000)
```

```python
def getMembers(groups):
    members = []
    for i in range(10):
        result = groups.results[i]
        temp = client.GetMembers(group_id=result['id'])
        for m in temp.results:
            members.append(m)
    
    return members
```

```python
def noEventsVSBioLength(group_list):
    x = []
    y = []
    
    for groups in group_list:
        for i in range(200):
            try:
                group = groups.results[i]
                group_id = group['id']
                bio_len = 0
                bio_count = 0
                members = client.GetMembers(group_id = group_id)
                for m in members.results:
                    if 'bio' in m:
                        bio_len = bio_len + len(m['bio'].split())
                        bio_count = bio_count + 1
                events = client.GetEvents(group_id=group_id)
                no_events = len(events.results)
                if bio_count != 0 and no_events != 0:
                    x.append(no_events)
                    y.append(bio_len / bio_count)
            except Exception:
                pass
    return x, y
```

```python
# x, y = noEventsVSBioLength([tech_groups, art_groups, lgbt_groups, language_groups, business_groups, sport_groups])
x, y = noEventsVSBioLength([tech_groups, art_groups, lgbt_groups, language_groups, business_groups, sport_groups])
```

```python
plt.scatter(x, y,s = 5)
plt.xlim([0, 300])
plt.title("Avg bio length VS. No. of events")
plt.xlabel("No. of events")
plt.ylabel("Avg No. of words in bios")
plt
plt.show()
```
